|--------|-----|-----|-----|-----|
| A      | 100 | 900 | 0   | 0   |

With only 10% of the dataset being actual positives, we can easily say that the dataset is not balanced.  

```python
# Method TP FP FN TN
# A 100 900 0 0
# B 80 125 20 775
# C 25 25 75 875
# D 50 50 50 850
# E 0 0 100 900

# Create a table with the data
data = {
        'A': {'TP': 100, 'FP': 900, 'FN': 0, 'TN': 0},
        'B': {'TP': 80, 'FP': 125, 'FN': 20, 'TN': 775},
        'C': {'TP': 25, 'FP': 25, 'FN': 75, 'TN': 875},
        'D': {'TP': 50, 'FP': 50, 'FN': 50, 'TN': 850},
        'E': {'TP': 0, 'FP': 0, 'FN': 100, 'TN': 900}
}

```

### Metrics for Comparing Method Performance


1. **Precision**
   - **Represents**: The accuracy of positive predictions.
   - **Definition**: Precision is the ratio of true positives to the sum of true positives and false positives.
   - **Formula**: PR = TP / (TP + FP)  
   - **Range**: [0, 1]
   - **Interpretation**: Higher values are better, as they indicate fewer false positives among predicted positives.

2. **Recall** (also known as Sensitivity or True Positive Rate)
   - **Represents**: The ability to capture all actual positives.
   - **Definition**: Recall is the ratio of true positives to the sum of true positives and false negatives.
   - **Formula**: RC = TP / (TP + FN)
   - **Range**: [0, 1]
   - **Interpretation**: Higher values are better, as they indicate that the method is identifying most true positives.

3. **Specificity** (also known as True Negative Rate)
   - **Represents**: The ability to correctly identify negatives.
   - **Definition**: Specificity is the ratio of true negatives to the sum of true negatives and false positives.
   - **Formula**: SP = TN / (TN + FP)
   - **Range**: [0, 1]
   - **Interpretation**: Higher values are better, as they indicate that the method accurately identifies negatives with few false positives.

4. **False Negative Rate (FNR)**
   - **Represents**: The proportion of actual positives missed by the method.
   - **Definition**: FNR is the ratio of false negatives to the sum of true positives and false negatives.
   - **Formula**: FNR = FN / (TP + FN)
   - **Range**: [0, 1]
   - **Interpretation**: Lower values are better, as they indicate fewer missed positives.

5. **False Positive Rate (FPR)**
   - **Represents**: The proportion of actual negatives incorrectly identified as positives.
   - **Definition**: FPR is the ratio of false positives to the sum of false positives and true negatives.
   - **Formula**: FPR = FP / (FP + TN)
   - **Range**: [0, 1]
   - **Interpretation**: Lower values are better, as they indicate fewer false positives.

6. **Accuracy (ACC)**
   - **Represents**: The overall correctness of predictions.
   - **Definition**: Accuracy is the ratio of correctly predicted instances (true positives and true negatives) to the total number of instances.
   - **Formula**: ACC = (TN + TP) / (TP + FN + FP + TN)
   - **Range**: [0, 1]
   - **Interpretation**: Higher values are better, indicating a higher overall correctness.

7. **Spatial Accuracy**
   - **Represents**: The effectiveness of positive identification relative to all positive and false detections.
   - **Definition**: Spatial Accuracy (S) measures the ratio of true positives to the sum of true positives, false negatives, and false positives, offering a focused accuracy measure for identifying positives.
   - **Formula**: S =  TP / (TP + FN + FP)
   - **Range**: [0, 1]
   - **Interpretation**: Higher values are better, as they indicate a stronger positive detection performance with minimal false positives and negatives.

8. **F-measure (= F1 Score) (Fm)**
   - **Represents**: The harmonic mean of precision and recall, balancing both metrics.
   - **Definition**: F-measure combines precision and recall to give a single measure of a methodâ€™s effectiveness.
   - **Formula**: Fm = (2 * PR * RC) / (PR + RC) 
   - **Range**: [0, 1]
   - **Interpretation**: Higher values are better, as they indicate a balanced approach between identifying positives and avoiding false positives.

```python
# Precision (PR): TP / (TP + FP)
# Recall (RC): TP / (TP + FN)
# Specificity (SP): TN / (TN + FP)
# False negative rate (FNR): FN / (TP + FN)
# False Positive Rate (FPR): FP / (FP + TN)
# Accuracy (ACC): (TN + TP) / (TP + FN + FP + TN)
# Spatial accuracy (S) or Jaccard index (J): TP / (TP + FN + FP)
# F-measure (Fm): (2 * PR * RC) / (PR + RC)

# Calculate the metrics using the data for each method
metrics = {
    'A': {},
    'B': {},
    'C': {},
    'D': {},
    'E': {}
}

for method in data:
    TP = data[method]['TP']
    FP = data[method]['FP']
    FN = data[method]['FN']
    TN = data[method]['TN']

    # Precision
    PR = TP / (TP + FP) if (TP + FP) != 0 else 0

    # Recall
    RC = TP / (TP + FN) if (TP + FN) != 0 else 0

    # Specificity
    SP = TN / (TN + FP) if (TN + FP) != 0 else 0

    # False Negative Rate
    FNR = FN / (TP + FN) if (TP + FN) != 0 else 0

    # False Positive Rate
    FPR = FP / (FP + TN) if (FP + TN) != 0 else 0

    # Accuracy
    ACC = (TN + TP) / (TP + FN + FP + TN) if (TP + FN + FP + TN) != 0 else 0

    # Spatial Accuracy
    S = TP / (TP + FN + FP) if (TP + FN + FP) != 0 else 0

    # F-measure
    Fm = (2 * PR * RC) / (PR + RC) if (PR + RC) != 0 else 0

    metrics[method] = { 'PR': PR, 'RC': RC, 'SP': SP, 'FNR': FNR, 'FPR': FPR, 'ACC': ACC, 'S': S, 'Fm': Fm }

```

### Print out the results of every method

```python
for method in metrics:
    print(method)
    print('Precision: ', metrics[method]['PR'])
    print('Recall: ', metrics[method]['RC'])
    print('Specificity: ', metrics[method]['SP'])
    print('False negative rate: ', metrics[method]['FNR'])
    print('False positive rate: ', metrics[method]['FPR'])
    print('Accuracy: ', metrics[method]['ACC'])
    print('Spatial accuracy: ', metrics[method]['S'])
    print('F-measure: ', metrics[method]['Fm'])
    print('\n')
```
